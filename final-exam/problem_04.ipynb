### imports libraries

```python
import numpy as np
import pandas as pd
import random
```

### generate a data with 10000 random real numbers 

```python
numbers = np.random.randint(0, 10000, 10000)
```

### mean, median, standart deviation

```python
mean = float(numbers.mean())
median = np.median(numbers)
std_deviation = np.std(numbers)
```

### min and max values

```python
min_value = min(numbers)
max_value = max(numbers)

print(f"Min value: {min_value}")
print(f"Max value: {max_value}")
```

### total count of numbers that are greater than mean

```python
total_max_than_mean = len(numbers[numbers > mean])
print(f"Total Values that are greater than mean: {total_maxer_than_mean}")
```

### min-max scaling

```python
min_max_numbers = []
for i in range(len(numbers)):
    min_max_numbers.append(float((numbers[i] - min_value) / (max_value - min_value)))

print(min_max_numbers)
```
